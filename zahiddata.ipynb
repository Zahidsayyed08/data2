```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='energy',y='loudness',color='c').set(title='Energy vs Loudness')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='acousticness',y='popularity',color='b').set(title='Popularity vs acousticness')
```
